```python
import csv
import matplotlib.pyplot as plt 

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)

header = next(data)

result = []

for row in data:
    if row[4] != '':
        if row[0].split('-')[1] == "12":
            result.append(float(row[4]))

f.close()

plt.figure(figsize=(10,2))
plt.plot(result)
plt.xlabel('Day (December)')
plt.ylabel('The highnest temperature')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding="cp949")
data = csv.reader(f)

header = next(data)

mean = []
low = []
high = []

for row in data:
    if (row[2] != '') and (row[0] != '') and (row[4] != ''):
        if(row[0].split('-')[1] == '12') and (row[0].split('-')[2] == '25'):
            mean.append(float(row[2]))
            low.append(float(row[3]))
            high.append(float(row[4]))

f.close()

plt.figure(figsize=(10, 2))
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title('크리스마스 기온 변화 그래프')
plt.plot(mean, 'k', label='mean')
plt.plot(low, 'b', label='low')
plt.plot(high, 'r', label='high')
plt.xlabel('Day (December 25)')
plt.ylabel('Temperature')
plt.legend()
plt.grid()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding="cp949")
data = csv.reader(f)

header = next(data)

temp_day = []
temp_mean = []
year = []

for row in data:
    if int(row[0].split('-')[0]) > 1907:
        if row[2] != '':
            temp_day.append(float(row[2]))
            if (row[0].split('-')[1] == '12') and (row[0].split('-')[2] == "31"):
                year.append(int(row[0].split('-')[0]))

                temp_mean.append(sum(temp_day) / len(temp_day))
                temp_day = []

f.close()

plt.figure(figsize=(10, 2))
plt.rc('font', family="Malgun Gothic")
plt.title('평균 기본 변화 그래프')
plt.plot(year, temp_mean, color='k', marker='d')
plt.xlabel('Year')
plt.ylabel('Average Temperauture')
plt.grid()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)
jan = []
aug = []

for row in data:
    month = row[0].split('-')[1]
    if row[4] != '':
        if month == '01':
            jan.append(float(row[4]))
        elif month == '08':
            aug.append(float(row[4]))

plt.hist(jan, bins=100, color='b', label='Jan')
plt.hist(aug, bins=100, color='r', label='Aug')
plt.legend()
plt.show()
```
